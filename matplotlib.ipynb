```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()  # Create a figure
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data 
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

# left, bottom, width, height (range 0 to 1)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'bo')    

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots()  # Create a figure and an axes
ax.plot(x, x, label='linear')  # Plot some data on the axes
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes
ax.plot(x, x**3, label='cubic')  
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
#implicit axes

x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  
plt.plot(x, x**2, label='quadratic')  
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(1, 1, 1)
x = np.linspace(0, 2, 100)
y = x**2
y2 = x**4
y3 = x**5
ax1.plot(x, y, color = 'red', linewidth = 4, label = ' $y = x^2$ ', zorder = 2) # zorder 2 = la retta rossa va sopra
ax1.plot(x, y2, color = 'blue', linewidth = 4, label = ' $y = x^4$ ', zorder = 1)
ax1.plot(x, y3, color = 'green', linewidth = 4, label = ' $y = x^5$  ', zorder = 0)

ax1.set_xlabel('x', fontsize = 26, loc = 'right')
ax1.set_ylabel('y', fontsize = 26, loc = 'top')

ax1.tick_params(axis = 'both', which = 'major', labelsize = 24, direction = 'in', length = 10) # major ticks
ax1.tick_params(axis = 'both', which = 'minor', labelsize = 24, direction = 'in', length = 3) # minor ticks 
ax1.set_xticks(ticks = ax1.get_xticks(), minor = True)
ax1.set_yticks(ticks = ax1.get_yticks(), minor = True)
ax1.minorticks_on()
ax1.set_title("Plot") 


ax1.set_xlim(left = 0, right = 2)
ax1.set_ylim(bottom = 0, top = 15)
ax1.legend(prop = {'size': 28}, ncol = 1)
plt.show()
```
