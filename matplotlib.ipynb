```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()  # Create a figure
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data 
```

```python
x = np.linspace(0, 5, 10)

y = x ** 2
fig = plt.figure()

# left, bottom, width, height (range 0 to 1)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y, 'bo')    

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots()  # Create a figure and an axes
ax.plot(x, x, label='linear')  # Plot some data on the axes
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes
ax.plot(x, x**3, label='cubic')  
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
#implicit axes

x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')  
plt.plot(x, x**2, label='quadratic')  
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python

fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(1, 1, 1)
x = np.linspace(0, 2, 100)
y = x**2
y2 = x**4
y3 = x**5
ax1.plot(x, y, color = 'red', linewidth = 4, label = ' $y = x^2$ ', zorder = 2) # zorder 2 = la retta rossa va sopra
ax1.plot(x, y2, color = 'blue', linewidth = 4, label = ' $y = x^4$ ', zorder = 1)
ax1.plot(x, y3, color = 'green', linewidth = 4, label = ' $y = x^5$  ', zorder = 0)

ax1.set_xlabel('x', fontsize = 26, loc = 'right')
ax1.set_ylabel('y', fontsize = 26, loc = 'top')

ax1.tick_params(axis = 'both', which = 'major', labelsize = 24, direction = 'in', length = 10) # major ticks
ax1.set_title("Plot") 


ax1.set_xlim(left = 0, right = 2)
ax1.set_ylim(bottom = 0, top = 15, )
ax1.legend(prop = {'size': 28}, ncol = 1, loc=2) # loc 2  in alto a sinistra
plt.show()
```

```python
# line customization
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(1, 1, 1)
x = np.linspace(0, 5, 50)
y = x**2
y2 = x**4
y3 = x**5


# possible linestype options -, , -., :
# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', 
# '1', '2', '3', '4', ...

ax1.plot(x, y, color = 'black', linewidth = 2, linestyle='-', marker='o', markersize=4, markerfacecolor="red", markeredgecolor="red",label = ' $y = x^2$ ', zorder = 2) # zorder 2 = la retta rossa va sopra
ax1.plot(x, y2, color = 'blue', linewidth = 2, linestyle='-.', marker='s', markersize=6,label = ' $y = x^4$ ', zorder = 1)
ax1.plot(x, y3, color = 'green', linewidth = 1, linestyle=':',label = ' $y = x^5$  ', zorder = 0)

ax1.set_xlabel('x', fontsize = 26, loc = 'right')
ax1.set_ylabel('y', fontsize = 26, loc = 'top')

ax1.tick_params(axis = 'both', which = 'major', labelsize = 22, direction = 'in', length = 10) # major ticks
ax1.tick_params(axis = 'both', which = 'minor', labelsize = 22, direction = 'in', length = 3) # minor ticks 

ax1.minorticks_on()
ax1.set_title("Plot") 


ax1.set_xlim(left = 0, right = 5)
ax1.set_ylim(bottom = 0, top = 20, )
ax1.legend(prop = {'size': 28}, ncol = 1, loc=2) # loc 2  in alto a sinistra
plt.show()
```

```python
# histograms
n = np.random.randn(1000000)
fig, axes = plt.subplots(1, 2, figsize=(16,6))

axes[0].hist(n, bins = 18, color = 'blue')
axes[0].set_title("Default histogram")
axes[0].set_xlim((min(n), max(n)))

axes[1].hist(n, cumulative=True, bins=70, color = 'red')
axes[1].set_title("Cumulative histogram")
axes[1].set_xlim((min(n), max(n)));
axes[1].text(0.65, 0.3, r"helloo", fontsize=20, color="blue")
```

```python

x = np.random.randn(1000,3)
n_bins = 10

fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2, figsize=(10,6))

colors = ['blue', '#e9c2f2', '#b729c2']
ax0.hist(x, n_bins, density=True, histtype='bar', color=colors, label=colors)
ax0.legend(prop={'size': 10})
ax0.set_title('bars with legend')

ax1.hist(x, n_bins, density=True, histtype='bar', color=colors,stacked=True)
ax1.set_title('stacked bar')

ax2.hist(x, n_bins, histtype='step', stacked=True,color=colors, fill=False)
ax2.set_title('stack step (unfilled)')

# Make a multiple-histogram of data-sets with different length.
x_multi = [np.random.randn(n) for n in [10000, 5000, 6464]]
ax3.hist(x_multi, n_bins, histtype='bar', color=colors)
ax3.set_title('different sample sizes')

fig.tight_layout()
plt.show()
```
